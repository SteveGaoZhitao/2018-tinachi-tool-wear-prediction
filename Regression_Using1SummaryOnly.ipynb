```python
#简单线性回归模型

import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import train_test_split

import pandas as pd
import matplotlib.pyplot as plt

import os.path
```

```python
srcpath='H:\\12_Kaggle\\toollife\\pythonProject\\PreProcessed'

srcfilename='SensorAllSummary3.csv'

srcfile=os.path.join(srcpath,srcfilename)
```

```python
#读入数据的行
df_x_train=pd.read_csv(srcfile) 
#需要去掉最后一行，第一2列
df_x_train=df_x_train.iloc[0:-1,2:]
#转换为np格式
x=np.array(df_x_train)
```

```python
df_x_train.columns
```

```python
ylist=[x for x in range(1,37)]
ylist.reverse()
y=np.array(ylist)
```

```python
X_train, X_test, y_train, y_test = train_test_split(x,y,test_size=0.33)
```

```python
len(X_train),len(X_test),len(y_train),len(y_test)
```

```python
reg=LinearRegression().fit(X_train,y_train)
```

```python
reg.score(X_train,y_train),reg.score(X_test,y_test),reg.coef_,reg.intercept_
```

```python
y_pred=reg.predict(X_test)
```

```python
X_test
```

```python
# Plot outputs
plt.plot(y_test,  color='black',label='real')
plt.plot(y_pred, color='blue', linewidth=3,label='predicted')

plt.xticks(())
plt.yticks(())
plt.title("")
plt.legend()

```

```python
# Plot outputs
plt.plot(y_train,  color='black',label='real_train')
plt.plot(reg.predict(X_train), color='blue', linewidth=3,label='predicted_train')

plt.xticks(())
plt.yticks(())
plt.title("")
plt.legend(loc=4)# 4代表右下角
```
